# Activation Function
* sigmoid (logistic function)
* tanh
* Relu
* leaky Relu

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-5,5,100)
```

```python
def myplot(x,y,title):
    fig = plt.figure(figsize=(12, 3))
    ax = fig.gca()
    plt.grid(b=True, which='major', color='#666666', linestyle='-')
    plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
    plt.plot(x,y,linewidth=2)
    plt.xlim(-5, 5)
    ax.set_ylabel('y')
    ax.set_xlabel('x')
    ax.set_title(title)
    plt.show()
```

# Sigmoid

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def dsigmoid(y):
    return y*(1-y)
```

```python
myplot(x,sigmoid(x),'sigmoid')
```

```python
y=sigmoid(x)
dy=dsigmoid(y)
myplot(x,dy,'dsigmoid')
```

# tanh

```python
def tanh(x):
    return np.tanh(x)

def dtanh(y):
    return 1-y*y
```

```python
myplot(x, tanh(x),'tanh')
```

```python
y=tanh(x)
dy=dtanh(y)
myplot(x,dy,'dtanh')
```

# Relu (Rectified Linear Unit)

```python
def relu(x):
    y=x.copy()
    y[y<0]=0
    return y

def drelu(y):
    a=y.copy()
    a[a>0]=1
    return a
```

The Rectified Linear Unit is a linear function without negative output, defined by the following equation

$y=\left\{\begin{matrix} if x>0: x\\otherwise: 0\end{matrix}\right.$

```python
myplot(x, relu(x),'RelU')
```

```python
y=relu(x)
dy=drelu(y)
myplot(x,dy,'dtanh')
```
