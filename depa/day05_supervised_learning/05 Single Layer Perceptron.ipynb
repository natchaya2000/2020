```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
```

```python
n=200
def normal(n):
    return np.random.randn(n)

x1=np.concatenate((normal(n)-2,normal(n)+2),axis=None)
x2=np.concatenate((normal(n)-2,normal(n)+2),axis=None)
yt=np.ones(2*n)
yt[:n]=-1
plt.plot(x1[yt==-1],x2[yt==-1],'or')
plt.plot(x1[yt==1],x2[yt==1],'ob')
```

```python
import numpy as np
from matplotlib import pyplot as plt
step = 0.1
mx2, mx1 = np.mgrid[slice(-6,6, step),
                    slice(-6,6, step)]

theta1 = 1.0 #w1
theta2 = 0.0 #w2
theta3 = 1.0 #bias
alpha =0.1
x=np.linspace(-6,5,10)
def func_yp(x1,x2):
    global theta1,theta2,theta3
    z=theta1*x1 + theta2*x2 +theta3
    return np.tanh(z)

plt.plot(x1[yt==-1],x2[yt==-1],'or')
plt.plot(x1[yt==1],x2[yt==1],'ob')
plt.xlim(-6, 6)
plt.ylim(-6, 6)
for i in range(100):
    yp=func_yp(x1,x2)
    theta1 = theta1 - alpha*np.average( (yp-yt)*(1-yp*yp)*x1 )
    theta2 = theta2 - alpha*np.average( (yp-yt)*(1-yp*yp)*x2 )
    theta3 = theta3 - alpha*np.average( (yp-yt)*(1-yp*yp ))
    if i%10==0:
        #print(theta1,theta2,theta3,)
        plt.plot(x, (-theta3-theta1*x)/theta2 ,'b')
        print('loss: %.3f'%np.mean((yp-yt)*(yp-yt)))
cmap = plt.get_cmap('PuOr')
plt.pcolormesh(mx1, mx2, func_yp(mx1,mx2), cmap=cmap)
```
