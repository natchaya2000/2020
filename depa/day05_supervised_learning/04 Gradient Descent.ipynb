# Gradient Descent

Gradient Descent Method is a numerical method to find optimum sulotion, where the the parameter $\theta$ makes minimum output of the subject function $L(\theta)$ by updating the previous $\theta$ using the gradient(slope)

$$\theta_{t+1} = \theta_{t} - \alpha \frac {dL(\theta)}{d\theta}$$

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

# Define Subject function

Let
$$f(x)=x^2-x-2$$

and its derivative is

$$\frac{\partial f}{\partial x}=2x-1$$

```python
def f(x):
    return np.power(x,2)-x+2

def df(x):
    return 2*x-1
```

# Show the Graph

```python
fig = plt.figure(figsize=(10,5))
ax = fig.gca()
plt.xlim(-5, 5)
plt.ylim(0, 30)
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
ax.set_ylabel(r'L($\theta$)')
ax.set_xlabel(r'$\theta$')
ax.set_title(r'$L(\theta)=\theta^2$')

theta=np.linspace(-5,5,50)
plt.plot(theta, f(theta),linewidth=3)
plt.show()
```

# Learning rate $\alpha = 0.1$

```python
alpha=0.1
theta_o=-4

fig = plt.figure(figsize=(10,5))
ax = fig.gca()
plt.xlim(-5, 5)
plt.ylim(0, 30)
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
ax.set_ylabel(r'L($\theta$)')
ax.set_xlabel(r'$\theta$')
ax.set_title(r'$L(\theta)=\theta^2, \alpha=%.3f, \theta=%.3f$'%(alpha,theta_o))

theta=np.linspace(-5,5,50)
ax.plot(theta,f(theta),linewidth=3)

theta_i=theta_o
Ls=[f(theta_i)]
thetas=[theta_i]
for i in range(20):
    theta_i=theta_i-alpha*df(theta_i)
    Ls.append(f(theta_i))
    thetas.append(theta_i)
ax=plt.plot(thetas,Ls,'or--') 
```
